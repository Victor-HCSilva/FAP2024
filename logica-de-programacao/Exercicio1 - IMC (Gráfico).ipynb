# Classe IMC

```python
import matplotlib.pyplot as plt

class Imc:
    def __init__(self, peso=63.00, altura=1.67, idade=30):
        self.peso = peso
        self.altura = altura
        self.idade = idade
        self.imc = 0
        self.mensagem = ''
        self.condicao = ''

    def calculo(self):
        '''Calcula o IMC'''
        self.imc = float(self.peso / self.altura**2)
        return self.imc

    def categoria(self):
        '''Define a categoria de IMC baseada na idade'''
        if self.idade < 60:
            if self.imc < 18.5:
                self.mensagem = 'Precisa de calorias e nutrientes.'
                self.condicao = 'abaixo do peso.'
            elif self.imc < 25:
                self.mensagem = 'Muito bem!'
                self.condicao = 'peso normal.'
            elif self.imc < 30:
                self.mensagem = 'Você precisa de uma dieta.'
                self.condicao = 'sobrepeso.'
            elif self.imc < 35:
                self.mensagem = 'Cuidado!'
                self.condicao = 'obesidade grau I.'
            elif self.imc < 40:
                self.mensagem = 'Muito cuidado!'
                self.condicao = 'obesidade grau II.'
            else:
                self.mensagem = 'Vá para o hospital!'
                self.condicao = 'obesidade mórbida (grau III).'
        elif self.idade >= 60:
            if self.imc < 22:
                self.mensagem = 'Precisa de calorias e nutrientes.'
                self.condicao = 'abaixo do peso.'
            elif self.imc < 27:
                self.mensagem = 'Muito bem!'
                self.condicao = 'peso normal.'
            elif self.imc < 30:
                self.mensagem = 'Você precisa de uma dieta.'
                self.condicao = 'sobrepeso.'
            elif self.imc < 35:
                self.mensagem = 'Cuidado!'
                self.condicao = 'obesidade grau I.'
            elif self.imc < 40:
                self.mensagem = 'Muito cuidado!'
                self.condicao = 'obesidade grau II.'
            else:
                self.mensagem = 'Vá para o hospital!'
                self.condicao = 'obesidade mórbida (grau III).'
        else:
            self.mensagem = 'Idade inválida.'
            self.condicao = ''

    def resultado(self):
        '''Retorna o resultado do IMC e a mensagem correspondente'''
        self.calculo()
        self.categoria()
        if self.mensagem != '':
            return f'{self.mensagem} Seu IMC é {self.imc:.1f}, que é classificado como {self.condicao}' 
        else:
            return 'Dados inválidos.'
```

# Loop

```python
# Lista para armazenar os dados dos participantes
participantes = []
while True:
    try:
        nome = input('Qual o seu nome?\n')
        peso = float(input("Qual é seu peso em Kg (Ex: 63.2)?\n"))
        altura = float(input("Qual é sua altura em metros (Ex: 1.67)?\n"))
        idade = float(input('Qual a sua idade (Ex: 23)?\n'))

        
        if 0 < idade < 120 and peso > 15 and peso < 250 and altura > 0.6 and altura < 2.5:
            imc = Imc(peso, altura, idade)
            print(imc.resultado())

            #dados do participante
            participantes.append((nome, imc.imc, altura, peso, idade))
        else:
            print('\nDados inválidos.')

        mais_participantes = input('Mais algum participante? (s/n): ')
        if mais_participantes.lower() != 's':
            break

    except Exception as e:
        print(f'\nDado inválido: {e}')
```

# Gráfico

```python
# Separando os dados para os gráficos
BLUE = (0, 0, 1)
nomes = [p[0] for p in participantes]
imcs = [p[1] for p in participantes]
alturas = [p[2] for p in participantes]
pesos = [p[3] for p in participantes]
idades = [p[4] for p in participantes]

# Calculando a média dos IMCs
imc_media = sum(imcs) / len(imcs)

# Gráfico 1 - IMC dos participantes
font1 = {'family':'serif','color':(0,0,0.5),'size':15}
font2 = {'family':'serif','color':(0,0.2,0),'size':15}
font3 = {'family':'serif','color':(0.5,0,0),'size':20}

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.suptitle('Análise do IMC', fontdict = font3)
plt.bar(nomes, imcs, color=BLUE, width = 0.4)
plt.axhline(imc_media, color='red', linestyle='--', label=f'Média: {imc_media:.2f}')
plt.title('IMC dos participantes', fontdict = font2)
plt.xlabel('Nomes', fontdict = font1)
plt.ylabel('IMC', fontdict = font1)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid()

# Gráfico 2 - Relação entre peso e idade
plt.subplot(1, 2, 2)
plt.scatter(idades, pesos, alpha=0.7, marker = 'o', lw = 10, color= (1,0.5,0))
plt.title('Relação entre peso e idade', fontdict = font2)
plt.xlabel('Idade', fontdict = font1)
plt.ylabel('Peso (kg)', fontdict = font1)
plt.grid()
plt.tight_layout()
plt.show()
```
